### Extract Tweets Based on Hashtags, Keywords, or User Accounts

```python
# Define the hashtag and the number of tweets to extract
hashtag = "#health"
num_tweets = 20  # Number of tweets to extract

# Use Tweepy Cursor to fetch tweets
tweets = tweepy.Cursor(api.search_tweets, q=hashtag, lang="en").items(num_tweets)

# Store tweets in a list
tweets_data = [[tweet.id_str, tweet.created_at, tweet.text, tweet.user.screen_name, tweet.favorite_count, tweet.retweet_count] for tweet in tweets]

# Convert the list to a DataFrame
df = pd.DataFrame(tweets_data, columns=['Tweet ID', 'Created At', 'Text', 'User', 'Likes', 'Retweets'])
```
